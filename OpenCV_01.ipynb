```python
# MATPLOT lib reads in Red Green Blue, but in this case it reads as Blue Green Red
fix_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # notice the capital C in cvtColor
```

```python
plt.imshow(fix_image)
```

```python
## Resizing and image
squeezed_img = cv2.resize(fix_image, (1000,400)) # 400 us related to 1300 (height) and 1000 is to 1900 ( width) kind of opposite
```

```python
plt.imshow(squeezed_img)
```

```python
# Reducing the size of image by ratio
w_ratio = 0.5
h_ratio = 0.5
```

```python
ratio_resized_image = cv2.resize(fix_image,(0,0),fix_image, w_ratio,h_ratio)
```

```python
plt.imshow(ratio_resized_image)
```

```python
ratio_resized_image.shape
```

```python
## Flipping image , horizontal axis = 0, vertical axis. = 1
flipped_image1 = cv2.flip(ratio_resized_image,0)
```

```python
plt.imshow(flipped_image1)
```

```python
flipped_image2 = cv2.flip(ratio_resized_image,1)
```

```python
plt.imshow(flipped_image2)
```

```python
flipped_image3 = cv2.flip(ratio_resized_image, -1)
```

```python
plt.imshow(flipped_image3)
```
